```python
# 3. visualizing the survival rate of class 1 
class_1 = []
count = 0 
for i in range (891):
  if (read.Pclass[i]==1):
    count += 1 
    class_1.append(read.Survived[i])
total_1 = sum(class_1)
print("Total number of people in class 1 are: {}" .format(count))
print("Total number of people survived in class 1 are {} " .format(sum(class_1)))
print("percentage of number of people survived in the class 1 are {}" .format(total_1 * 100/count))
```

```python
# 7. Model Building - Linear Regression
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

X = read.iloc[:, 0].values.reshape(-1, 1)  
Y = read.iloc[:, 1].values.reshape(-1, 1)  
linear_regressor = LinearRegression()  
linear_regressor.fit(X, Y)  
Y_pred = linear_regressor.predict(X)  

plt.scatter(X, Y)
plt.plot(X, Y_pred, color='purple')
plt.show()
```
